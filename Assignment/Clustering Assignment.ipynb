### compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference.

```python
import numpy as np
from sklearn.cluster import KMeans
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = scipy.misc.face(gray=True)
plt.figure(figsize=(10, 3.6))
plt.imshow(f, cmap=plt.cm.gray)
plt.show()
```

```python
f.shape
```

```python
X = f.reshape(f.shape[0]*f.shape[1], 1)
```

```python
kmeans = KMeans(n_clusters=5).fit(X)
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(f.shape)
```

```python
segmented_img
```

```python
#saving image
plt.imsave('compressed_rac.png',segmented_img);
```

```python
# Visualize the compressed image

image = plt.imread('compressed_rac.png')
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.show()
```
